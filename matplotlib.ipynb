## Import matplotlib library 

```python
import matplotlib.pyplot as plt    
```

```python
x = [1,2,3,4]                   # creating data for visualize Line plot/ line chart 
y = [20,25,23,30]
```

#### Line chart/ Line Plot

```python
#line chart/ Line Plot
plt.plot(x, y)                           # X axis = horizontol  And  Y axis = verticals axis 
plt.show()
```

```python
#Crearting BaKary Cake sales Per Week
x = ["Mon","Tue","Wed","Thur","Friday"]  #x Axis
y = [11,15,8,10,12]                      #y Axis

plt.plot(x,y)          #line plot function

plt.title("Bakery Sales This Week!")       #title of graph
  
plt.xlabel("Days of the Week")                   #X axis
plt.ylabel("Sales of Per Day Cakes")             # Y axis 

plt.show()                            
```

#### Eight Important Function Of Matplotlib 

```python
#Parameterized Plot 
#plt.plot(x,y, color='name', linestyle='lineStyle', linewidth = value, marker ='marker symbol' ,label='label name')
#plt.xlabel("text")
#plt.ylabel("text")

months = [1,2,3,4]
sales = [1000,1500,1100,1700]

plt.figure(figsize=(8,5))
plt.plot(months,sales, color="Blue", linestyle="--", linewidth = 2, marker ='o',label="2025 Sales Data")
plt.xlabel("Months")
plt.ylabel("Sales Per Month")
plt.title("Monthly Sales Reports")
plt.legend(loc="upper left")                                   #showing lable
plt.grid(color='gray',linestyle=':',linewidth=1)               #modifying graph
plt.xlim(1,4)                            #limitize X Axis
plt.ylim(0,2000)                         #limitize Y Axis
plt.xticks([1,2,3,4],["M1","M2","M3","M4"])
plt.show()
```

## Bar Plot

```python
#syntx of bar plot
#plt.bar(x,height, color="name", width=value, label='labelname')
#plt.barh()   for horizontale bar chat 
#if data is categorize  bar chart Approch is best 
products = ["A","B","C","D"]
sales = [1000,1500,800,1200] 


plt.bar(products,sales, color='Red', label='Sales 2025')
plt.xlabel("products")
plt.ylabel("sales")
plt.title("Products Sales Comparison")
plt.legend()
plt.show()
```

## Pie Plot

```python
#syntax
#plt.pie(values, label=labels_list,colors=colors_list, autopct='%1.1f%%',shadow=True)


regions = ["North","South","East","West"]
revenue = [3000,2000,1500,1100]
plt.figure(figsize=(7,5))
plt.pie(revenue, labels=regions,autopct="%1.1f%%",colors=["gold","lightgreen","skyblue","coral"],shadow=True)
plt.title("Revenue Contribution by Region")
plt.show()
```

## Histograms 

```python
# Histograms Syntax 
#Best Approch is making hiistogram when my data is Continous......
#plt.hist(data, bins=Numbers_of_bins, color='colorname', edgecolor='black')......  

scores = [48,70,55,66,75,70,88,77,83,56,78,72,64,50,86,62,68]

plt.hist(scores, bins=5, color="darkblue", edgecolor="black")
plt.xlabel("Score Range")
plt.ylabel("Numbers of Student")
plt.title("Score Distribution Of Students",color="darkBlue")
plt.show()
```

## Scatter Plot 

```python
#Scatter Plot Syntax                                   # x axis euals to input and y axis = output .... like linear equation ot findin Co-relation
#plt.scatter(x,y, color='color_name', marker='markerStyle', label='label_name')

studied_hours = [1,2,3,4,5,6,7,8]
exam_scores  =  [50,55,60,65,70,75,80,85]

plt.scatter(studied_hours,exam_scores, color='green', marker='o', label='Student Data')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Relationship Between Study Time And Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Comparing two or multiple group/data Sets in single graph

plt.scatter([1,2,3],[55,60,65], color='red', label='class A') #group 1
plt.scatter([1,2,3],[52,57,60], color='Blue', label='class B') #group 2

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Comparison of two classes")
plt.legend()
plt.grid(True)
plt.show()
```

## Subplot & Layout Adjustments

```python
#Subplot Syntax                     best things to make  Subplot mutliple charts in single figure and Quickly analyze related data sets together..
#plt.subplot(nrows, ncols, index)

x = [1,2,3,4]
y = [10,20,15,23]

plt.figure(figsize=(10,3))
plt.subplot(1,2,1)     # 1st_row, 2column, 1st subplot 
plt.plot(x,y)
plt.title("Line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar chart")

#plt.tight_layout()

plt.show()
```

```python
#Advanced/Professional method 
#fig, ax = plt.subplot(nrow, ncols, figsize=(width, height)) 
x = [1,2,3,4]
y = [10,20,15,23]

fig, ax= plt.subplots(1,2, figsize= (10,5))

ax[0].plot(x,y, color="red")
ax[0].set_title("Line chart")

ax[1].bar(x,y,color="darkblue")
ax[1].set_title("Bar chart")

fig.suptitle("Comparison of Line And Bar Charts")

plt.tight_layout()           #Automatic Adjustment Overlap fix... 
plt.show()
```
